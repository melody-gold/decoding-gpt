```python
import numpy as np
from traitlets import Float
```

```python
# check square array
# accept int of words to generate in vocab
def generate_sequence(vocab, initial, numWords: int = 5):
    # given initial word, generate sequence of words
    # check for square vocab matrix
    if vocab.shape[0] != vocab.shape[1]:
        raise ValueError("Vocab matrix must be square")
    
    current_word = initial
    sequence = [current_word]
    for _ in range(numWords):
        next_word_probs = vocab[current_word]
        next_word = np.random.choice(len(vocab), p=next_word_probs)
        sequence.append(next_word)
        current_word = next_word
    return sequence
    
```

```python
# generate a sequence of states
cheese_vocab: list[str] = ["I", "like", "cheese"]
cheese_chain: Float[np.ndarray, "d_vocab.d_vocab"] = np.array([
	[0.5, 0.5, 0],
	[0, 0, 1],
	[1, 0, 0],
], dtype=np.float64)

initial_state: str = "I"
initial_index: int = cheese_vocab.index(initial_state)
generated_indices = generate_sequence(cheese_chain, initial_index, numWords=10)
generated_words = [cheese_vocab[i] for i in generated_indices]
print("Generated sequence:", " ".join(generated_words))
```

```python
def generate(
    chain: Float[np.ndarray, "d_vocabd_vocab"],
    vocab: list[str],
    start_state: str,
    max_length: int = 10,
) -> None:
    assert start_state in vocab, "Start state must be in vocabulary"
    assert chain.sum(axis=1).all() == 1.0, "Each row of the chain must sum to 1"
    d_vocab: int = len(vocab)
    assert tuple(chain.shape) == (d_vocab, d_vocab), "Chain must be of shape (d_vocab, d_vocab)"
    output: list[str] = [start_state]
    
    for _ in range(max_length):
        current_idx: int = vocab.index(output[-1])
        next_state: int = np.random.choice(d_vocab, p=chain[current_idx])
        output.append(vocab[next_state])
    
    return print(" ".join(output))

generate(chain=cheese_chain, vocab=cheese_vocab, start_state="I", max_length=10)
```
